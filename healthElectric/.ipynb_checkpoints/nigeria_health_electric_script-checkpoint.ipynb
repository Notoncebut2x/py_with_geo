```python
# Import the necessary packages
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gpd
import rasterio
from rasterio.plot import show
import fiona
import matplotlib.pyplot as plt
import rioxarray as rxr
from shapely import wkt
from shapely.geometry import Point
```

```python
# Import data

# Import Electrical Distribution Grid tif (via facebook)
tif_file = r'./data/electrical_grid_nigeria_15_electrical_distribution_grid.tif/electrical_grid_nigeria_15_electrical_distribution_grid.tif'
raster = rasterio.open(tif_file)

# Import health facility vector point data (via healthsites.io)
healthsites = gpd.read_file("./data/nigeria-shapefiles/shapefiles/healthsites.shp")
coords = healthsites.geometry
```

```python
# Check if imports worked

#show(raster)
display(healthsites)

# Check CRS of the files
print(healthsites.crs)
print(raster.crs)

# Dispaly the points over the raster
fig, ax = plt.subplots(figsize=(12,12))
healthsites.plot(ax=ax, color='orangered')
show(raster, ax=ax)

```

```python
# Plot histogram of electric fig
raster2 = rxr.open_rasterio('./data/electrical_grid_nigeria_15_electrical_distribution_grid.tif/electrical_grid_nigeria_15_electrical_distribution_grid.tif', masked=True)


f, ax = plt.subplots(figsize=(10, 6))
raster2.plot.hist(ax=ax,
       color="purple")
ax.set(title="Distribution of Lidar DEM Elevation Values",
       xlabel='Elevation (meters)',
       ylabel='Frequency')
plt.show()
```

```python
#Extact xy of healthsites points
for point in healthsites['geometry']:
    print(point.xy[0][0],point.xy[1][0])

    
for point in healthsites['geometry']:
    x = point.xy[0][0]
    y = point.xy[1][0]
    row, col = raster.index(x,y)
    print("Point correspond to row, col: %d, %d"%(row,col))
    #print("Raster value on point %.2f \n"%raster.read(1)[row,col])
```

```python
values = pd.Series()

# Read input shapefile with fiona and iterate over each feature
with fiona.open("./data/nigeria-shapefiles/shapefiles/healthsites.shp") as shp:
    for feature in shp:
        osm_id = feature['properties']['osm_id']
        coords = feature['geometry']['coordinates']
        # Read pixel value at the given coordinates using Rasterio
        # NB: `sample()` returns an iterable of ndarrays.
        with rasterio.open('./data/electrical_grid_nigeria_15_electrical_distribution_grid.tif/electrical_grid_nigeria_15_electrical_distribution_grid.tif') as src:
            value = [v for v in src.sample([coords])][0][0]
        # Update the pandas serie accordingly
        values.loc[osm_id] = value
values.view()     
        
# Write records into a CSV file
values.to_csv('data/out.csv')
```

```python
# Join the values based on osm_id to coordinates

# Read the data we just created
healthsites_mv = pd.read_csv('data/out.csv')
#healthsites_mv.head()

#Lets give the healthsites_mv more readable attibute names
healthsites_mv = healthsites_mv.rename(columns = {'Unnamed: 0':'osm_id', '0':'Electricity '})
#healthsites_mv.head()



#Subet the healthsite data for the attributes we are interested in
healthsites_geo = healthsites[['osm_id','geometry']]
#healthsites_geo.head()

# Perform an Attribute Join to merge the datasets together
healthsites_join = healthsites_geo.merge(healthsites_mv, on='osm_id')

#healthsites_join.plot()
#healthsites_join.head()

# Export the data to map
healthsites_join.to_file("data/healthsites_electric.gpkg", driver="GPKG")

```
